```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
from nltk.corpus import stopwords
import nltk
from keras.utils.np_utils import to_categorical
nltk.download('stopwords')
nltk.download('wordnet')
df = pd.read_csv("https://raw.githubusercontent.com/garyongguanjie/moviecsv/master/movie_dataset.csv")
print(df.shape)
df = df.drop_duplicates(subset='movie_title', keep="first")
# df['rating'] = df['rating'].apply(lambda x:"safe" if(x=="PG13"or x=="PG") else "unsafe")
print(df.shape)
df['movie_title']
```

## Clean Texts
### Remove punctuation,stopwords,"\n"
### Convert to lower case
### Lemmatize text
### remove numbers

```
import re
RE_D = re.compile('\d')
def containsNum(string):
    return RE_D.search(string)
```

```
stop = set(stopwords.words('english'))
df['synopsis_clean'] = df['synopsis'].str.replace('[{}]'.format(string.punctuation), ' ').replace('\n'," ").str.lower()
df['synopsis_clean']= df['synopsis_clean'].apply(lambda x: ' '.join([item for item in x.split() if item not in stop]))
lemmatizer = nltk.stem.WordNetLemmatizer() 
df['synopsis_clean']= df['synopsis_clean'].apply(lambda x: ' '.join([lemmatizer.lemmatize(item) for item in x.split()]))
df['synopsis_clean']= df['synopsis_clean'].apply(lambda x: ' '.join([item for item in x.split() if not containsNum(item) ]))
```
